```python
import vaex as vx
import h5py
import sys
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl



vx_df = vx.open('/Users/khanh/UCSF/npr/raw_data/npr_mlsmr.hdf5') #change me
# grep only npr values                 
npr_df = vx_df.to_pandas_df(["npr1", "npr2"])
mpl.rcParams['figure.dpi'] = 500

x = npr_df['npr1']
y = npr_df['npr2']
print('Plotting...')
# plot data with 1000x1000    , set color bar to base 10, color scheme  
plt.hist2d(x, y, (1000, 1000), norm=mpl.colors.LogNorm(), cmap=plt.cm.get_cmap('Spectral_r'))
cb = plt.colorbar()
cb.set_label('Number of molecules')
plt.clim(1, 100000)
plt.ylim((0.5, 1.02))
plt.xlim((0, 1))
plt.title("NPR heatmap")
plt.xlabel('npr1')
plt.ylabel('npr2')

```
